```python
############################
### import modules

import sys
import uproot
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math
import csv
import io
import matplotlib.colors as colors
from scipy.optimize import curve_fit
from scipy.stats import pearsonr
import astropy
from astropy import stats
from scipy import interpolate
import os

###########################

np.set_printoptions(suppress=True)
os.chdir("../data")  # moves working directory to data folder
```

```python
tau = 480 # mean lifetime of semi-stable states in units of nanoseconds

#mucap = 1e-10 # rate of neutron radiative capture
#musca = 100*mucap # rate of neutron inelastic scatter, Fermi estimation

x = np.arange(0, 2000, 1)

def eff(x, tau):
    eff = 1 - np.exp(- x/tau)
    return eff

#effback = (1 - np.exp(- mucap*x))*(1 - np.exp(- musca*x))

plt.figure(figsize=(14,12))

plt.plot(x, eff(x, 480), color="blue", label=r'$\tau = 480$ ns')
plt.plot(x, eff(x, 200), color="red", label=r'$\tau = 200$ ns')
plt.plot(x, eff(x, 1000), color="green", label=r'$\tau = 1e3$ ns')
plt.plot(x, eff(x, 1e7), color="orange", label=r'$\tau = 1e7$ ns')
#plt.plot(x, effback, color="red", label='Neutron signal coincidence')

plt.xlabel('Limiting time [ns]',fontsize=30)
plt.xticks(fontsize=28)
plt.ylabel(r'Probability decay',fontsize=30)
plt.yticks(fontsize=28)

plt.legend(fontsize=28)
plt.savefig('../figures/limittime.png', bbox_inches='tight')

plt.show()
```
